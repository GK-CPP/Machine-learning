```python
import numpy as np
'''
training_data=[
    [10,10],
    [12,10],
    [22,10],
    [38,10],
    [40,9],
    [44,9],
    [45,8],
    [48,8],
    [51,7],
    [54,7],
    [55,6],
    [58,6],
    [59,6],
    [60,5],
    [63,5],
    [66,4],
    [68,4],
    [71,3],
    [74,3],
    [77,2],
    [78,2],
    [84,1],
    [88,1],
    [92,1]
]
'''
training_data = [ [17,1],
                 [42,2],
                 [38,1],
                 [25,1],
                 [44,2],
                 [47,3],
                 [49,3],
                 [50,4],
                 [54,4],
                 [53,4]
                ]

print(training_data)
dec={}
arr=[]
x_ans=[]
```

```python
def data(att):
        arr.clear()
        dec.clear()
        for i in att:
            dec[i[1]]=0;
        for i in att:
                if(dec[i[1]]==0):
                    arr.append(i[1])
                dec[i[1]]+=1
        #print(arr)
       # for i in arr:
          #  print(dec[i])
sum=0
```

```python
def gini(att):
    sum=0
    total_d=len(att)
    for i in arr:
        sum+=((dec[i]/float(total_d))**2)
    return (1-sum)
```

```python
data(training_data)
total_gini=gini(training_data)
print(total_gini)
```

```python
tree1=[[]]
tree2=[[]]
def tree_data(att,indx):
    tree1.clear()
    tree2.clear()
    for i in att:
        if(i[0]<=indx[0]):
            tree1.append(i)
        else:
            tree2.append(i)

def IG(tr1,tr2,g1,g2):
    l1=len(tr1)
    l2=len(tr2)
    total_len=l1+l2;
    Ig=total_gini-((l1/float(total_len))*g1)-((l2/float(total_len))*g2)
    return Ig
```

```python
IG_arr=[[]]
ma=0
ans_1=ans_2=0
for i in training_data:
    tree_data(training_data,i)
    print(tree1)
    print(tree2)
    data(tree1)
    per_gini1=gini(tree1)
    data(tree2)
    per_gini2=gini(tree2)
    Ig=IG(tree1,tree2,per_gini1,per_gini2)
    IG_arr.append([Ig,i[0]])
    if(Ig>ma):
        ma=Ig
        ans_1=(i[0])
        ans_2=(i[1])
    print(per_gini1)
    print(per_gini2)
    print(Ig)
    print("\n")
print(IG_arr)
print(ans_1,ans_2)
x_ans.append([ans_1,ans_2])
```

```python
sss=[ans_1,ans_2]
tree_data(training_data,sss)

training_data.clear()
for i in tree2:
    training_data.append(i)
print(training_data);
total_gini=gini(training_data)
print(total_gini)
IG_arr.clear()
ma=0
ans_1=ans_2=0
for i in training_data:
    tree_data(training_data,i)
    print(tree1)
    print(tree2)
    data(tree1)
    per_gini1=gini(tree1)
    data(tree2)
    per_gini2=gini(tree2)
    Ig=IG(tree1,tree2,per_gini1,per_gini2)
    IG_arr.append([Ig,i[0]])
    if(Ig>ma):
        ma=Ig
        ans_1=(i[0])
        ans_2=(i[1])
    print(per_gini1)
    print(per_gini2)
    print(Ig)
    print("\n")
print(IG_arr)
print(ans_1,ans_2)
x_ans.append([ans_1,ans_2])
```
